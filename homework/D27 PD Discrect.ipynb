### 丟一個銅板，丟了100次，出現正面 50 次的機率有多大。
(提示: 先想是哪一種分配，然後透過 python 語法進行計算)


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
```

```python
n_trial = 100
p = 0.5
x = np.linspace(0,n_trial,n_trial+1)
probs = stats.binom.pmf(x, n_trial, p)
plt.bar(x, probs)
plt.ylabel('P(X=x)')
plt.xlabel('x')
plt.title('binomial(n={0},p={1})'.format(n_trial,p))
plt.show()
```

```python
cumsum_probs = stats.binom.cdf(x, n_trial, p)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title('binomial(n={0},p={1})'.format(n_trial,p))
plt.plot(x, cumsum_probs)
plt.show()
```

```python
X = stats.binom.rvs(n_trial,p,size=5)
print(X)
plt.hist(X)
plt.show()
```
