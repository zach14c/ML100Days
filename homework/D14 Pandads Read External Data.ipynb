### 2. 請將 Dcard API 取得所有的看板資訊轉換成 DataFrame，並且依照熱門程度排序後存成一個 csv 的檔案。


```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums')
response = r.text

import json
data = json.loads(response)
```

```python
df_f = pd.json_normalize(data)
df_f.head(5)
```

```python
df_f.sort_values(by='subscriptionCount',inplace=True,ascending=False)
```

```python
df_f.head(5)
```

```python
df_f.to_csv('dcard_f.csv')
%ls -l dcard_f.csv
```
