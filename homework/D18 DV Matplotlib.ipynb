# 範例重點

如何使用亂數, 資料集來操作


```python
# 載入需要的...
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math

# 準備數據 ... 假設我要畫一個sin波 從0~180度
# 設定要畫的的x,y數據list....
x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)
# 開始畫圖
plt.plot(x,y)
plt.show()

#畫出完整的 sin 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
plt.plot(x, y_sin)
plt.show()

#畫出完整的 cos 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)
plt.plot(x, y_cos)
plt.show()

```

# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

'''
#作業
#方法不只有一種
'''
def F(X, Y):
    c = np.zeros(len(X))
    for i in range(0,len(X)):
        if X[i] > 0:
            c[i] = math.atan(Y[i]/X[i])
        elif X[i] < 0 and Y[i] >= 0:
            c[i] = math.atan(Y[i]/X[i])+math.pi
        elif X[i] < 0 and Y[i] < 0:
            c[i] = math.atan(Y[i]/X[i])-math.pi            
        else:
            c[i] = math.pi/2
    return c
        

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.scatter(X,Y, s=100, alpha=0.3, c=F(X,Y))
```
